# PART 3B: COMPARE WITH XGB MODEL

XGBoost models have consistently turned in strong results on such tasks, so let's see if it can do better than the log-reg version in notebook3.0, or even the Distilbert model.

The process is exactly the same as in notebook3.0, with changes accounting for XGB's parameters and the use of RandomizedSearchCV for the sake of time.

```python
import itertools
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import joblib 

from sklearn.dummy import DummyClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

mpl.rcParams["figure.dpi"] = 300
%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

```python
raw = pd.read_csv("../data/train_raw.csv")

val = pd.read_csv("../data/validate.csv").dropna(subset=["clean_text"])
```

```python
X = raw["clean_text"]
y = raw["troll_or_not"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2, shuffle=True, stratify=y
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

# 1. DUMMY BASELINE

The model will have to better than 0.5, which is what you get with a random guess.

```python
dummy = DummyClassifier(strategy='stratified', random_state=42)
```

```python
dummy.fit(X_train, y_train)
dummy_pred = dummy.predict(X_test)
print("Accuracy Score:", accuracy_score(y_test, dummy_pred))
```

# 2. GRID SEARCH FOR OPTIMAL PARAMETERS

Same BOW approach. Switch up the parameters range if you have more powerful machines at your disposal. This ran for nearly 8 hours on my late-2015 iMac, even using a random search CV instead of the usual gridsearch. 

```python
model = Pipeline(
    [
        ("vect", CountVectorizer(stop_words="english", ngram_range=(1, 1))),
        ("tfidf", TfidfTransformer(norm="l2", use_idf=True)),
        (
            "clf",
            XGBClassifier(
                base_score=0.5,
                objective="binary:logistic",
                n_estimators=1000,
                n_jobs=-1,
                verbosity=1,
            ),
        ),
    ]
)

parameters = {
    "vect__min_df": [i / 1000 for i in range(1, 5)],
    "clf__learning_rate": [0.0001, 0.001, 0.01, 0.1],
    "clf__seed": [16, 32, 64],
    "clf__reg_lambda": [i for i in range(1, 10)],
    "clf__gamma": [i / 10.0 for i in range(1, 10)],
    "clf__reg_alpha": [1e-5, 1e-2, 0.1, 1, 10],
    "clf__max_depth": [i for i in range(5, 30)],
    "clf__max_delta_step": [i for i in range(1, 10)],
    "clf__min_child_weight": [i for i in range(1, 10)],
    "clf__gamma": [i / 10.0 for i in range(1, 10)],
    "clf__reg_alpha": [1e-5, 1e-2, 0.1, 1, 10],
    "clf__colsample_bylevel": [i / 10.0 for i in range(1, 10)],
    "clf__colsample_bytree": [i / 10.0 for i in range(1, 10)],
    "clf__colsample_bynode": [i / 10.0 for i in range(1, 10)],
    "clf__subsample": [i / 10.0 for i in range(1, 10)],
}
```

```python
# RandomizedSearchCV used instead of GridSearchCV due to the large number of parameters tuned
# RandomizedSearchCV completes in a far shorter time, with a slight dent on performance
# See https://scikit-learn.org/stable/auto_examples/model_selection/plot_randomized_search.html

gs_clf_xgb = RandomizedSearchCV(
    model,
    parameters,
    cv=5,
    n_jobs=-1,
    n_iter=1000,
    random_state=16,
    scoring="accuracy",
    verbose=1,
    return_train_score=True,
)

gs_clf_xgb = gs_clf_xgb.fit(X_train, y_train)

print("Best Score:", gs_clf_xgb.best_score_)
print("Best Parameters:", gs_clf_xgb.best_params_)
```

```python
XGB = joblib.load("../pkl/xgb_rdcv.pkl")
```

```python
# Showing the finalised model the full training data-set, ahead of testing it on unseen data
# This takes a long time to load on a laptop due to the file size

XGB.fit(X, y)
```

## 3.1 METRICS

The XGB model's precision score on the validation set is slightly lower than the log-reg version. The XGB model scores slightly higher for accuracy, recall and F1 - but not enough to beat the fine tuned Distilbert model.  

```python
# Loading up validation set

X_validate = val['clean_text']
y_validate = val['troll_or_not']
```

```python
pred_validate = XGB.predict(X_validate)
```

```python
cm_validate = confusion_matrix(y_validate, pred_validate)
cm_validate
```

```python
print("Accuracy score: ", accuracy_score(y_validate, pred_validate))

print("Precision score: ", precision_score(y_validate, pred_validate))

print("Recall score: ", recall_score(y_validate, pred_validate))

print("F1 score: ", f1_score(y_validate, pred_validate))
```

## 3.2: CONFUSION MATRIX

This gives us a better look at how the XGB model performed on the validation set. What jumps out right away is the higher number of false positives - real tweets that the model predicted as troll tweets - compared to both the log-reg model as well as the Distilbert model.

While it did pick out more troll tweets than the log-reg model, the high number of false positives is a concern.

```python
plt.figure(figsize=(12,12))
sns.heatmap(cm_validate, annot=True, fmt="d", cmap="gist_gray_r", annot_kws={"size": 20})
plt.title("Predictions v Actual", fontsize=20)
plt.ylabel("Actual ", fontsize=20)
plt.xlabel("Predicted ", fontsize=20)
plt.tight_layout()
```

# 4. TEST LOG-REG MODEL AGAINST IRAN DATASET

Like the Distilbert model and the log-reg version, this XGB model was trained on state troll tweets from China and Russia. How will it perform when tasked with classifing state troll tweets from Iran?

```python
# this dataset is in the repo

iran = pd.read_csv("../data/iran.csv")
```

```python
cols1 = ["tweetid", "bot_or_not", "user_screen_name", "tweet_text", "clean_tweet_text"]

iran = iran[cols1].copy()

iran = iran.rename(
    columns={
        "tweetid": "tweetid",
        "bot_or_not": "troll_or_not",
        "user_screen_name": "user_display_name",
        "tweet_text": "tweet_text",
        "clean_tweet_text": "clean_text",
    }
)

iran = iran.dropna(subset=['clean_text']).copy()
```

```python
# Loading up Iran set

X_iran = iran['clean_text']
y_iran = iran['troll_or_not']
```

```python
pred_iran = XGB.predict(X_iran)
```

## 4.1: METRICS

All the metrics dropped significantly across the board, though not as badly as the log-reg version.

```python
print("Accuracy score: ", accuracy_score(y_iran, pred_iran))

print("Precision score: ", precision_score(y_iran, pred_iran))

print("Recall score: ", recall_score(y_iran, pred_iran))

print("F1 scoe: ", f1_score(y_iran, pred_iran))
```

## 4.2: CONFUSION MATRIX FOR IRAN DATASET 

When exposed to state troll tweets from another country, the XGB model similarly fails to perform as well, resulting in a very high number of false negatives. It only succeeded in correctly picking out about half the troll tweets (532 out of 1,000) in this dataset.

The fine tuned Distilbert model held up much better in this scenario, raising hopes that it might do as well against, say, state troll tweets from Saudi Arabia or Bangladesh. 

```python
cm_iran = confusion_matrix(y_iran, pred_iran)
cm_iran
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(cm_iran, annot=True, fmt="d", cmap="gist_gray_r", annot_kws={"size": 20})
plt.title("Predictions v Actual", fontsize=20)
plt.ylabel("Actual ", fontsize=20)
plt.xlabel("Predicted ", fontsize=20)
plt.tight_layout()
```
